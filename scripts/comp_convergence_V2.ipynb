```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import time
from IPython import display
import os
```

```python
def comp_plot(path_dict):
    
#     base_psnr = 31.65
    fig, axs = plt.subplots(1, 1, figsize=(15, 10.5))
#     axs.axhline(y=base_psnr, linestyle='--')

        
    for key, values in path_dict.items():
        df = pd.read_csv(values)
        df_psnr = df['psnr']
        
        if key == 'baseline':
            num_epoch_base = len(df) 
        if key == 'ours':
            axs.axhline(y=df_psnr.values[-1], linestyle='--')
            axs.axvline(x=len(df_psnr)-1, linestyle='--')
        axs.plot(df_psnr[:-1], label=key, linewidth=2.5)
#         sns.lineplot(data=df_psnr, palette="tab10", linewidth=2.5)        
    num_epoch_ours = len(df)
    axs.legend()
    plt.xticks(range(0, 301, 20))
    plt.yticks(range(25, 31, 1))
    axs.grid(linestyle='--')
    plt.show()
    is_complete = True if num_epoch_base==num_epoch_ours else False
    return is_complete
    
path_dict = {
    'baseline': '../experiments/SRDENSENETD16C430.16/records/train_records.csv',
#     'Improve-1': '../experiments/EDENSENET_V1_woRes/records/train_records.csv',
  #  'Improve-1+Res': '../experiments/EDENSENET_V1_wRes/records/train_records.csv',
#     'Improve-3': '../experiments/EDENSENET_V3_woCA/records/train_records.csv',
   # 'V1+LocR': '../experiments/EDENSENET_V1+locR/records/train_records.csv',
#     'Imporve1+CA':'../experiments/EDENSENET_V1+CA/records/train_records.csv',
#     'Imporve-1+locR+CA': '../experiments/EDENSENET_in3f64_x4/records/train_records.csv'
 #   'V1+LocR+w': '../experiments/EDENSENET_in3f64_x4/records/train_records.csv'
   # 'compress2v0': '../experiments/COMPRESSV229.85/records/train_records.csv',
    #'V1+LocR_RA': '../experiments/EDENSENET_V1+locR_relu_after/records/train_records.csv',
    #'V1+locR+RA+W': '../experiments/EDENSENET_in3f64_x4/records/train_records.csv'
    #'compress2': '../experiments/COMPRESSV229.85_relu_after/records/train_records.csv',
    #'compress2wca': '../experiments/COMPRESSV2_in3f64_x4/records/train_records.csv',
    #'V1localRreluafter': '../experiments/EDENSENET_V1+locR_relu_after/records/train_records.csv',
    #'+updown':'/home/server606/clh/SCS_IRSR/experiments/EDENSELAP30.05/records/train_records.csv',
    #'+spl': '/home/server606/clh/SCS_IRSR/experiments/EDENSEUPDOWN_in3f64_x4/records/train_records.csv',
    'edense': '../experiments/EDENSEUPDOWN_in3f64_x4/records/train_records.csv'
}
result_path = path_dict['edense']
ctime = int(os.stat(result_path).st_ctime)
refresh_time = 10
count = 1

while True:
    if ctime==int(os.stat(result_path).st_ctime) and count!=1:
        time.sleep(refresh_time)
    else:
        is_complete = comp_plot(path_dict)
        print('NO.%d refresh | Refresh Time: ' %count, time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time())))
        count += 1
        display.clear_output(wait=True)
        ctime = int(os.stat(result_path).st_ctime)
        if is_complete: break
```
